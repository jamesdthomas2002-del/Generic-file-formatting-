```python
import pandas as pd 

FILE_PATH = r"C:\Useful Join Files\Dealer names.csv"
FILTER_COL = 'Dealer Name'

df = pd.read_csv(FILE_PATH, lineterminator = '\n', encoding = 'utf-8-sig')
df = df.drop(columns=['\r'])
df[FILTER_COL]= df[FILTER_COL].str.strip()
df[FILTER_COL] = df[FILTER_COL].astype(str)
df['Dealer Code'] = df[FILTER_COL].str[-4:]
df['Dealer Code'] = df['Dealer Code'].str.strip(')')
df[FILTER_COL] = df[FILTER_COL].str[:-7]

print(df.head())


#df.to_csv(r"C:\Useful Join Files\Dealer names and Codes.csv")

```
